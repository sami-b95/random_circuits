## Plot XEB benchmark

```python
def plot_1d_xeb(ax, legend, df, **kwargs):
    ### Generate mean and error plots
    df_selection = None
    for field, value in kwargs.items():
        if df_selection is None:
            df_selection = (df[field] == value)
        else:
            df_selection &= (df[field] == value)
    selected_df = df[df_selection]
    plot = ax.errorbar(selected_df["length"], selected_df["xeb_mean"], yerr=selected_df["xeb_std"], linestyle="-", marker=".")
    plot.set_label(legend)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_yscale("log")
plot_1d_xeb(ax, "Moment 1", df.query("length > 5"), n_qubits=8, moment=1)
plot_1d_xeb(ax, "Moment 2", df.query("length > 5"), n_qubits=8, moment=3)
plot_1d_xeb(ax, "Moment 3", df.query("length > 5"), n_qubits=8, moment=5)
ax.set_xlabel("Circuit length")
ax.set_ylabel("Benchmark")
ax.set_title("Generalized XEB benchmark for 1D circuit", y=1.05)
ax.legend()
```

### Plot XEB benchmarks as heatmap

```python
def plot_2d_xeb_heatmap(df, s_range, c_range, **kwargs):
    ### Generate mean and error maps
    mean_map = np.empty((len(s_range), len(c_range)), dtype=float)
    error_map = np.empty((len(s_range), len(c_range)), dtype=float)
    for s_index, s in enumerate(s_range):
        for c_index, c in enumerate(c_range):
            df_selection = (df["s"] == s) & (df["c"] == c)
            for field, value in kwargs.items():
                df_selection &= (df[field] == value)
            row = df[df_selection]
            mean_map[s_index, c_index] = row["xeb_mean"]
            error_map[s_index, c_index] = row["rel_xeb_std"]
    ### Plot them
    fig = plt.figure(figsize=(15, 5))
    # Mean map
    ax1 = fig.add_subplot(1, 2, 1)
    im1 = ax1.imshow(mean_map[:, ::-1].T, cmap=cm.coolwarm, extent=[s_range.start, s_range.stop, c_range.start, c_range.stop])
    ax1.set_xticks(range(s_range.start, s_range.stop, s_range.step))
    ax1.set_yticks(range(c_range.start, c_range.stop, c_range.step))
    ax1.set_yticks(list(c_range))
    ax1.set_xlabel("$s$")
    ax1.set_ylabel("$c$")
    ax1.set_title("Benchmark", y=1.05)
    fig.colorbar(im1, ax=ax1)
    # Error map
    ax2 = fig.add_subplot(1, 2, 2)
    im2 = ax2.imshow(error_map[:, ::-1].T, cmap=cm.coolwarm, extent=[s_range.start, s_range.stop, c_range.start, c_range.stop])
    ax2.set_xticks(range(s_range.start, s_range.stop, s_range.step))
    ax2.set_yticks(range(c_range.start, c_range.stop, c_range.step))
    ax2.set_yticks(list(c_range))
    ax2.set_xlabel("$s$")
    ax2.set_ylabel("$c$")
    ax2.set_title("Relative error", y=1.05)
    fig.colorbar(im2, ax=ax2)
    plt.show()
```

```python
plot_2d_xeb_heatmap(
    df,
    s_range=range(1, 4),
    c_range=range(1, 4),
    moment=1,
    n_qubits_per_dimension=4
)
```
